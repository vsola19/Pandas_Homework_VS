```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School file and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
# Read Student file and store into Pandas DataFrames
student_data_df = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
#Calculate the percentage of students with a passing math score >= 70 (6) - answer appears at datatype: float64
students_passing_math = student_data_df.loc[student_data_df["math_score"] >= 70, ["math_score"]].count()
students_passing_math
percentage_passed_math = (students_passing_math/school_total_students) * 100 
percentage_passed_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater) (7) - answer appears as datatype: float64
students_passing_reading = student_data_df.loc[student_data_df["reading_score"] >= 70, ["reading_score"]].count()
students_passing_reading
percentage_passed_reading = (students_passing_reading/school_total_students) * 100 
percentage_passed_reading
```

```python
#Calculate the percentage of students with a passing of both math and reading score (70 and greater)
overall_passing = student_data_df[(student_data_df['math_score'] >=70) & 
                (student_data_df['reading_score']>=70)]['student_name'].count()

overall_passing 
```

```python
#Calculate the percentage of students with a passing of both math and reading score (70 and greater)
overall_passing = (student_data_df[(student_data_df['math_score'] >=70) & 
                (student_data_df['reading_score']>=70)]['student_name'].count()/school_total_students) * 100

overall_passing 
```

```python
#Resetting of index as data type error was being produced.
percentage_passed_math.reset_index(inplace = True, drop = True)
percentage_passed_reading.reset_index(inplace = True, drop = True)
```

```python
#Formatting applied to number of students and budget
district_summary = {
    "Total Schools": total_schools,
    "Total Number of Students":"{:,}".format(school_total_students),
    "Total Budget": "${:,.2f}".format(school_total_budget),
    "Average Math Score": average_math_score,
    "Average Read Score": average_reading_score,
    "% Passing Math": percentage_passed_math,
    "% Passing Reading": percentage_passed_reading,
    "% Overall Passing": overall_passing}
#How to create data frame:
district_summary_df = pd.DataFrame(district_summary)
district_summary_df
```
